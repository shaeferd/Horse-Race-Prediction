```python
def plot_graphs(data):
    #histogram 
    fig, axs = plt.subplots(2,2, figsize = (13,13));
    hist = sns.distplot(data, bins= 30, kde=False, ax = axs[0][0]);
    hist.set_title('Histogram');

    #lag
    lag = data.copy()
    lag = np.array(lag[:-1])
    current = data[1:]

    ax = sns.regplot(current,lag,fit_reg=False, ax = axs[0][1]);
    ax.set_ylabel("y_i-1");
    ax.set_xlabel("y_i");
    ax.set_title('Lag');

    #QQ
    qntls, xr = stats.probplot(data, fit=False);
    qq = sns.regplot(xr,qntls, ax = axs[1][0]);
    qq.set_title('QQ');


    #Run Sequence
    ax = sns.regplot(np.arange(len(data)),data,fit_reg=False);
    ax.set_ylabel("val")
    ax.set_xlabel("i")
    ax.set_title('Run Sequence')
```

```python
df_races_norm['horse_name'] = df_races['horse_name']
df_races_norm
```

```python
df_byhorse = df_races_norm.groupby('horse_name').mean()
df_byhorse.head()
```

```python
pca_model = skd.PCA().fit(df_races_norm.drop(['horse_name'], axis = 1))
pca_model.components_.shape
```

```python
pca_model.explained_variance_
```

```python
plt.plot(range(1,13),pca_model.explained_variance_,'b-o')
```

```python
#Keep 4 components
```

```python
# Scatterplot of disaggregated data
X = pca_model.transform(df_races_norm.ix[:,:12])
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA biplot')
plt.ylim(-4, 6)

# Add variable unit vector projections
V = pca_model.transform(np.identity(X.shape[1]))
for i, v in enumerate(V):
    plt.annotate(df_races_norm.columns[i], 
                 xy=(0,0), xytext=v[:2]*10, 
                 fontsize=13, color='orange',
                 arrowprops=dict(
                    arrowstyle='<-', linewidth=2, color='orange'))
```

```python
# Scatterplot of aggregated data
X = pca_model.transform(df_byhorse)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], s=50)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA: Aggregated Data\nPerceptual Map')
plt.xlim(-6,6)
plt.ylim(-4,6)

# Add variable unit vector projections
V = pca_model.transform(np.identity(X.shape[1]))
for i, v in enumerate(V):
    plt.annotate(df_races_norm.columns[i], 
                 xy=(0,0), xytext=v[:2]*10, 
                 fontsize=12, color='orange',
                 arrowprops=dict(
                    arrowstyle='<-', linewidth=2, 
                    color='orange'))

# Add data point annotation
for i, s in enumerate(df_byhorse.index):
    plt.annotate(s, xy=X[i,:2]+[0.1,-0.1], fontsize=5)
```

```python
# Create a three-factor model
fa_model = skd.FactorAnalysis(n_components=2).fit(df_races_norm.drop(['horse_name'], axis = 1))

# Show the loadings
df_loadings = pd.DataFrame(fa_model.components_[:3,:].T, 
                   index=df_races_norm.describe().columns,
                   columns=['Factor1', 'Factor2'])
df_loadings
```

```python
# Create a three-factor model
fa_model = skd.FactorAnalysis(n_components=3).fit(df_races_norm.drop(['horse_name'], axis = 1))

# Show the loadings
df_loadings = pd.DataFrame(fa_model.components_[:3,:].T, 
                   index=df_races_norm.describe().columns,
                   columns=['Factor1', 'Factor2', 'Factor 3'])
df_loadings
```
